```python
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os
```

```python
online_retail = pd.read_excel('online_retail.xlsx')
```

```python
load_dotenv()
db_url = os.getenv("database_url")
```

Load the raw data into the SQL database using SQLAlchemy.

```python
engine = create_engine(db_url)
```

```python
online_retail.to_sql('online_retail_data', engine, if_exists='replace', index=False, schema='online_retail')
```
