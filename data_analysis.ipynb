### Country Performance (Excluding United Kingdom)  
Rank countries by total revenue  
Identify the top 10 revenue-generating countries  
Compute both revenue and quantity sold

```python
# Filter data to exclude uk
country_data = result[result['Country']!='United Kingdom']
```

```python
# Rank countries by total revenue (excluding uk)
country_revenue = country_data.groupby('Country')['Revenue'].sum().sort_values(ascending=False)
country_revenue
```

```python
# Rank top 10 countries by total revenue (excluding uk)
top_10_country_revenue = country_data.groupby('Country')['Revenue'].sum().sort_values(ascending=False).head(10)
top_10_country_revenue
```

```python
# Rank countries by total quantity sold(excluding uk)
country_quantity = country_data.groupby('Country')['Quantity'].sum().sort_values(ascending=False)
country_quantity
```

```python
# Rank top 10 countries by total quantity sold(excluding uk)
top_10_country_quantity = country_data.groupby('Country')['Quantity'].sum().sort_values(ascending=False).head(10)
top_10_country_quantity
```
